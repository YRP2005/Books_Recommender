#  Popularity based system

```python
rating_with_name = pd.merge(ratings,book,how='inner',on='ISBN')
```

```python
rating_with_name.sample(5)
```

```python
rating_with_name.groupby('Book-Title').count()
```

```python
rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

yahan basically groupby kiya and usse counts aaye yahan counts matlab is book ki entries mein kitne baar aaya vo data frame mein matlab
group by min jo rating hai na vo ye nahi ki kitne stars hain balki  ye hai ki kitne logon yaani kitne users ne rate kiya hai isi book ko

[['Book-Rating'] ka matlab ki ye jo df aayega naya group by karne ke baad usme kewal Book rating ka coloumn do along with title 
now groupby ke karan title hi index ban gaya hai isiliye  they want ki wapas se simple 1,2 ,3 ,4 index aajaye isiliye reset index

```python
num_rating=rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

yahan bas wahi rename kiya taaki  apne ko confusion na ho

```python
num_rating.rename(columns={'Book-Rating':'number_of_ratings'},inplace=True)
```

yahan basically ek naya data frame banya jisme sirf books title and unki mean ratings hai is baar count nahi ki kitni rating hai bas avergae rating

```python
average_ratings = rating_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
```

```python
average_ratings.rename(columns={'Book-Rating':'average_ratings'},inplace=True)
```

now we will merge avegrating df and nummber of rating df as usse hame ye milega book title,number of ratings,average rating 

```python
rating_full_info = pd.merge(average_ratings,num_rating,how="inner",on="Book-Title").reset_index()
```

now we ant books with highest number of ratings as they are the most reviewed and usme se highest avergae rating wali 100 nikal lenge as we want those ith highest number of reviews and good reviews

```python
rating_full_info[rating_full_info['number_of_ratings']>255]
```

```python
rating_full_info[rating_full_info['number_of_ratings']>255].sort_values('average_ratings', ascending=False)
```

till now we got book titles with more than 255 number of ratings and they are sorted according to thier average rating in descending order

now inme se top 100 nikalenge apan

```python
top_100_books= rating_full_info[rating_full_info['number_of_ratings']>255].sort_values('average_ratings', ascending=False).head(100)
```

```python
top_100_books.head()
```

```python
pd.merge(top_100_books,book,how="inner",on="Book-Title")
```

```python
top_100_books = pd.merge(top_100_books,book,how="inner",on="Book-Title")
```
